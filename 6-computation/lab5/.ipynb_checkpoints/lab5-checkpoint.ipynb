```python
Y = f(X)
plt.plot(X, Y, label="x tgx")
plt.scatter(x_nodes, y_nodes, label="intrp. nodes")
plt.legend()
```

# 1. Lagrange

```python
def calculate_lagrange(xs, ys):
    n = len(xs)
    C = []
    for k in range(0, n):
        c_k = 1
        for i in range(0, n):
            if i == k:
                continue
            c_k *= (xs[k] - xs[i])
        C.append( ys[k] / c_k )
        
    return C

def evaluate_lagrange(C, xs, ys, x, eps=1e-4):
    # d = np.abs(xs - x)
    # di, dm = np.argmin(d), np.min(d)
    # if dm < eps:
    #     return ys[di]
    
    n = len(C)
    y = 0
    for k in range(0,n):
        dx = 1
        for i in range(0, n):
            if i==k:
                continue
            dx *= (x - xs[i])
        y += C[k] * dx
    
    return y
    
    
```

```python
c1 = calculate_lagrange(x_nodes, y_nodes)

plt.plot(X, Y, label="true")

y1 = []
for x in X:
    y1.append(evaluate_lagrange(c1, x_nodes, y_nodes, x))

plt.scatter( x_nodes, y_nodes, label="nodes" )
plt.plot(X, y1, label="interp")
plt.title("Lagrange interpolation")
plt.legend()
```

```python
error1 = Y-y1
plt.plot(X, error1, label="error")
plt.axvline(x_nodes[0], lw=0.5, c='r', label='node')
plt.axhline(0, lw=0.5, c='k')
for x in x_nodes[1:]:
    plt.axvline(x, lw=0.5, c='r')
plt.legend()
plt.ylim(-0.003, 0.001)
plt.title("Lagrange interpolation error")
```

```python
max_error1 = np.abs(error1).max()
print("Max error:", max_error1)
```

### Теоретична похибка:
Let $M_n = \max_x |f^{(n+1)}(x)|$

$R_n(x) = \frac{M_n}{(n+1)!}(x-x_0)(x-x_1)...(x-x_n)$

$f^{(n+1)} = (x\tan x)^{(n+1)}$.

```python
t = sp.symbols('t')
sf = t * sp.tan(t)
sf1 = sp.lambdify(t, sp.diff(sf, t, N+1), "numpy")
plt.plot(X, sf1(X))
plt.gca().set_title("$(x tg x)^{(n+1)}$")
plt.ylim(-1e10, 1e10)
```

```python
def generateR1(n):
    
    M = sf1(np.array([-1,0,1])).max()
    k = 1/np.math.factorial(N+1)
    K = M * k
    print("Max sf=", M)
    print("M/k = ", K)
    
    def R1(x):
        out = K
        for node in x_nodes:
            out *= (x - node)
        return out
    return R1
        
error_fun1 = generateR1(N)
plt.plot(X, error_fun1(X), label="theoretical error")
plt.axvline(x_nodes[0], lw=0.5, c='r', label='node')
plt.axhline(0, lw=0.5, c='k')
for x in x_nodes[1:]:
    plt.axvline(x, lw=0.5, c='r')
plt.legend()
plt.ylim(-10, 10)
plt.title("Lagrange interpolation THEORETICAL error")
```

# 2. Newton

```python
def divdif(xs):
    if xs.shape[0] == 1:
        return f(xs)[0]
    
    if xs.shape[0] == 2:
        return (f(xs[1]) - f(xs[0])) / (xs[1] - xs[0])
    
    return (divdif(xs[1:]) - divdif(xs[:-1])) / (xs[-1] - xs[0])
```

```python
def calculate_newton(x1, y1):
    n = len(x1)
    C = []
    
    for k in range(1,n+1):
        C.append(divdif(x1[:k]))
        
    return C

def evaluate_newton(C, x1, y1, x, eps=1e-4):
    # d = np.abs(xs - x)
    # di, dm = np.argmin(d), np.min(d)
    # if dm < eps:
    #     return ys[di]
    
    n = len(C)
    y = C[0]
    dx = 1
    for k in range(0, n-1):
        dx *= (x - x1[k])
        y += C[k+1] * dx
        
    return y
```

```python
c2 = calculate_newton(x_nodes, y_nodes)
print(c2)
plt.plot(X, Y, label="true")

y2 = []
for x in X:
    y2.append(evaluate_newton(c2, x_nodes, y_nodes, x))

plt.scatter( x_nodes, y_nodes, label="nodes" )
plt.plot(X, y2, label="interp")
plt.legend()
plt.title("Newton interpolation")
```

```python
error2 = Y-y2
plt.plot(X, error2, label="error")
plt.axvline(x_nodes[0], lw=0.5, c='r', label='node')
plt.axhline(0, lw=0.5, c='k')
for x in x_nodes[1:]:
    plt.axvline(x, lw=0.5, c='r')
plt.legend()
plt.ylim(-0.003, 0.001)
plt.title("Newton interpolation error")
```

```python
max_error2 = np.abs(error2).max()
print("Max newton error:", max_error2)
```

## 3.1. Метод Томаса / прогонки
для ефективного розв'язку систем із тридіагональними матрицями

1. Визначення



2. Прямий хід




3. Зворонтій хід



---

Існує також більш лаконічний варіант алгоритму:


```python
# diag - main diagonal
# l_diag - lower subdiagonal
# u_diag - upper subdiagonal
def prohonka(diag, l_diag, u_diag, b_):
    n = b_.shape[0]
    
    a = l_diag.copy()
    b = diag.copy()
    c = u_diag.copy()
    d = b_.copy()
    
    for i in range(1, n):
        w = a[i-1] / b[i-1]
        b[i] = b[i] - w * c[i-1]
        d[i] = d[i] - w * d[i-1]
    
    x = np.zeros(n)
    x[-1] = d[-1] / b[-1]
    for i in range(n-2, -1, -1):
        x[i] = ( d[i] - c[i] * x[i+1] ) / b[i]
        
    return x

def test_prohonka():
    N = 1000
    err = 0
    merr = 0
    for i in range(N):
        d0 = (np.random.rand(3) - 0.5)
        d1 = (np.random.rand(4) - 0.5) * 3 
        d2 = (np.random.rand(3) - 0.5)

        A = np.diag(d0, -1) + np.diag(d1, 0) + np.diag(d2, +1)
        b = (np.random.rand(4)-0.5)
        x = prohonka(d1, d0, d2, b)
        
        eb = A@x - b
        e = np.linalg.norm(eb)
        err += e
        if e > merr:
            merr = e
        
        if i == 0:
            print("example matrices:")
            print("A:\n", A)
            print("b:", b)
            # print("x:", x)
            # print("x0:", np.linalg.inv(A) @ b)
            
        if e > 1e-8:
            print("A:", A)
            print("b:", b)
            print("x:", x)
            print("too large error:", e)
    
    print()
    print("Mean error:", err/N)
    print("Max error:", merr)
    
    
test_prohonka()
```

Це результат обчислення нев'язок при вирішенні 1000 випадкових тридіагональних систем методом прогонки
*Бачимо, що середня нев'язка достатньо мала, тобто алгоритм ймовірно реалізований правильно*

## 3.2. Власне сплайни

```python
def calculate_splines(x1, y1):
    n = x1.shape[0] - 1
    
    A = np.zeros((n,n))
    b = np.zeros(n)
    
    a = np.zeros(n)
    b = np.zeros(n)
    c = np.zeros(n)
    d = np.zeros(n)
    
    h = x1[1:] - x1[:-1]
    
    a = y1[:-1]
    
    l_diag = np.zeros(n-1)
    u_diag = np.zeros(n-1)
    diag = np.zeros(n)
    
    l_diag[:-1] = h[:-2]
    u_diag[1:] = h[2:]
    diag[1:-1] = (h[1:-1] + h[:-2]) * 2
    diag[0] = 1
    diag[-1] = 1
    
    print(l_diag.shape[0], u_diag.shape[0], diag.shape[0])
    
    B = np.zeros(n)
    B[1:-1] = 3 * ( (y1[2:-1] - y1[1:-2])/h[1:-1] - (y1[1:-2] - y1[:-3])/h[:-2] )
    B[0] = 0
    B[-1] = 0
    
    A = np.diag(diag) + np.diag(l_diag,-1) + np.diag(u_diag,+1)
    print("A:\n", A.round(2))
    print("b:", b.round(2))
     
    c = prohonka(diag, l_diag, u_diag, B)
    
    for j in range(n-1):
        d[j] = (c[j+1] - c[j]) / (3 * h[j])
        b[j] = ((y1[j+1] - y1[j])/h[j]) - (h[j]/3) * (c[j+1] + 2*c[j])
    d[-1] = -c[-1]/(3*h[-1])
    b[-1] = ((y1[-1] - y1[-2])/h[-1]) - (2 * h[-1] * c[-1]) / 3

    return a, b, c, d
```

```python

def eval_I(n, a, b, c, d, x1, y1, x):
    I = n
    for i in range(n):
        if x < x1[i]:
            I = i-1
            break
    return max(min(I, n-1), 0)

def evaluate_splines(n, a, b, c, d, x1, y1, x):
    I = eval_I(n, a, b, c, d, x1, y1, x)
    
    dx = x - x1[I]
    return a[I] + dx * (b[I] + dx * (c[I] + dx * d[I]) ) if I is not None else None
```

```python
n3 = x_nodes.shape[0]-1
print("n for splines = ", n3)

c3 = calculate_splines(x_nodes, y_nodes)
print("\na:\n", c3[0].round(2))
print("b:\n", c3[1].round(2))
print("c:\n", c3[2].round(2))
print("d:\n", c3[3].round(2))

plt.plot(X, Y, label="true")

y3 = []
Is = []
for x in X:
    y3.append(evaluate_splines(n3, *c3, x_nodes, y_nodes, x))
    Is.append(eval_I(n3, c3[0], c3[1], c3[2], c3[3], x_nodes, y_nodes, x)/10)
    
plt.scatter( x_nodes, y_nodes, label="nodes" )
plt.plot(X, y3, label="interp")
plt.plot(X, Is, label="I")
plt.legend()
plt.title("Spline interpolation")
```

```python
error3 = Y-y3
plt.plot(X, error3, label="error")
plt.axvline(x_nodes[0], lw=0.5, c='r', label='node')
plt.axhline(0, lw=0.5, c='k')
for x in x_nodes[1:]:
    plt.axvline(x, lw=0.5, c='r')
plt.legend()
plt.ylim(-0.1, 0.05)
plt.title("Spline interpolation error")
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
EPS = 1e-3
PREC = 4

A = np.array([
    [5.5, 7., 6., 5.5],
    [7., 10.5, 8., 7.],
    [6.0, 8.0, 10.5, 9.],
    [5.5, 7., 9., 10.5]
])
b = np.array([23, 32, 33, 31])

# доповнена матриця системи
Ab = np.hstack([A, b.reshape(-1,1)])
print(Ab)
```

$$Ax = b, \quad A = D + U + L$$, де $D$ - діагональна, $U$ - upper triangular, $L$ - lower triangular (тобто $U_{ii} = L_{ii} = 0$)

$$(D+U+L)x = b$$
$$(D+L)x = -Ux + b$$
$$x^{k+1} = -(D+L)^{-1}Ux^{k} + (D+L)^{-1}b = (D+L)^{-1}\left(b - Ux^{k}\right)$$

Тобто
$$C = -(D+L)^{-1}U, \quad d = (D+L)^{-1}b$$

Рівняння також можна записати як
$$(D+L)x = b - Ux$$

Тоді 
$$x_1 := \frac{1}{D_{11}} \left(b_1 - \sum_{j=2}^n U_{1j}x_j\right)$$
$$\begin{align*}x_2 &:= \frac{1}{D_{22}} \left(b_2 - D_{21}x_1 - \sum_{j=3}^n U_{2j}x_j\right)
= \frac{1}{D_{22}} \left(b_2 - \sum_{j=1}^n A_{2j}x_j + D_{22}x_2\right) = \\
&= x_2 + \frac{1}{D_{22}} \left(b_2 - \sum_{j=1}^n A_{2j}x_j \right)\end{align*}$$

Аналогічно
$$x_i := x_i + \frac{1}{D_{ii}} \left(b_i - \sum_{j=1}^n A_{ij}x_j \right)$$

Зручно обчислювати другий доданок-"поправку" окремо, бо він використовується у критерії зупинки

```python
def Seidel_StopCriterium():
    return True

def Seidel_Run(A, b, eps, timeout=40):
    n = b.shape[0]
    x = np.zeros(n)
    step = 0
    
    def Seidel_Step():
        max_delta = 0
        for i in range(n):
            delta = (b[i] - A[i,:].dot(x)) / A[i,i]
            delta = delta.round(PREC)
            x[i] = x[i] + delta
            if abs(delta) > max_delta:
                max_delta = abs(delta)
        return max_delta
    
    while True:
        nvzk = (b-A@x).round(PREC)
        print(f"i{step}: x={x}, нев'язка={nvzk} ({np.linalg.norm(nvzk).round(PREC)})")
        
        max_delta = Seidel_Step()
        if (max_delta < eps):
            print("Stopping by stop criterium")
            break
        if (step > timeout):
            print("Stopping by timeout")
            break
        step += 1
            
    nvzk = (b-A@x).round(PREC)
    print(f">   Result: x={x}, нев'язка={nvzk} ({np.linalg.norm(nvzk).round(PREC)})")
```

```python
Seidel_Run(A, b, EPS/10)
```

```python
print("Істинний корінь x* =", np.linalg.solve(A,b))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.stats
```

```python
X = np.array([
    [0.02, 0.02, 0.10, 0.10, 0.18, 0.18],
    [538, 649, 538, 649, 538, 649]
])
Y = np.array([
    78.9, 56.2, 80.9, 57.4, 85.3, 60.7
])
n = Y.shape[0]
```

```python
Z = np.vstack([np.ones(n), X])
A = Z @ Z.T
B = np.linalg.inv(A)
```

```python
w = B @ Z @ Y
w
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(20, 55)

ax.scatter3D(X[0], X[1], Y, c='r');

xmin = X.min(axis=1)
xmax = X.max(axis=1)
xs, ys = np.meshgrid(np.linspace(xmin[0], xmax[0], 10), np.linspace(xmin[1], xmax[1], 10))
ax.plot_surface(xs, ys, w[0] + xs*w[1] + ys*w[2], edgecolor='none', alpha=0.5)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def evaluate(k):
    t = sp.symbols('t')
    x = 0*t

    for i in range(8):
        x = t + 0.5 * x.subs(t, t**k)
        
    return sp.lambdify(t, x, "numpy"), x

for k in range(2,8):
    X = np.linspace(0,1,11)
    f, f_text = evaluate(k)
    Y = f(X)
    plt.plot(X, Y, label=f"k = {k}")
    plt.legend()
    plt.xticks(X)
```

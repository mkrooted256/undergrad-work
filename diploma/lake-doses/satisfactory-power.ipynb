```python
import scipy as sp
import numpy as np
import scipy.optimize

import matplotlib.pyplot as plt
```

```python
# Constructor
X1 = np.array([0.1, 0.2, 0.5, 1.0])
Y1 = np.array([0.2, 0.5, 1.6, 4.0])
plt.plot(X1, Y1, '-o', label='constructor')

# Assembler
X2 = np.array([0.1, 0.2, 0.5,  0.8,  1.0])
Y2 = np.array([0.7, 1.8, 6.0, 11.2, 15.0])
plt.plot(X2, Y2, '-o', label='assembler')

plt.legend()
```

```python
X = np.array([0.1, 0.2, 0.5,  0.8,  1.0])
Y = np.array([0.7, 1.8, 6.0, 11.2, 15.0]) / 15
```

```python
Xh = np.log(X)
Yh = np.log(Y)
```

```python
res = sp.stats.linregress(Xh, Yh, 'greater')
res
```

```python
plt.scatter(X, Y)
plt.plot(X, np.float_power(X, 1.328662))
```
